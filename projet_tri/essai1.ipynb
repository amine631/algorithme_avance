# Tri par Insertion:

```python
temps_d = datetime.now()
temps_triInsertion = []
tableau=[]

def triInsertion (v):
    p=len(v)-1
    for i in range(0,p) :
        x = v[i+1]
        j=i
        while((v[j] > x ) and (j>=0) ) :
            v[j+1] = v[j]
            j-= 1
        v[j+1] = x
    print("le tableau trié",v)
    temps_f = datetime.now()
    temps = temps_f-temps_d
    sec = temps.total_seconds()
    temps_triInsertion.append(sec)
    
    
    
    
borne_inf=800
borne_sup=5000
pas=500
for i in range(borne_inf, borne_sup+1, pas):
    tableau.append(i)
    a=rd.randint(100, size=(i))
    pas=len(a)-1
    triInsertion(a)
```
